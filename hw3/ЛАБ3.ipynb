```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from PIL import Image
```

```python
resnet = tf.keras.applications.ResNet50(weights='imagenet')
```

```python

image_path = '5259240.jpg'
image = Image.open(image_path)
image = image.resize((224, 224))
x = tf.keras.preprocessing.image.img_to_array(image)
x = tf.expand_dims(x, axis=0)
x = preprocess_input(x)

predictions = resnet.predict(x)
decoded_predictions = decode_predictions(predictions, top=1000)

image.show()

otter_probability = None
for obj in decoded_predictions[0]:
    if obj[1] == 'otter':
        otter_probability = obj[2]
        break

# Выводим вероятность, если класс 'buckeye' найден
if otter_probability is not None:
    print('Это ВЫДРА с вероятностью:', otter_probability)
else:
    print('Изображение не является выдрой вовсе')
```
